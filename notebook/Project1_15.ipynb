### Exercise 3, Part I: Finite Differences (FD) with Functions ###

#### part 1 ####

```python
# Import the package for data visualization
import matplotlib.pyplot as plt
```

```python
#function defined for wave phenomenon in python
def f(x, a=0.1, b=10):
    return np.sin(b*x)*np.exp(-a*x*x)
```

```python
# plot for f(x) from equation above for arbitary closed interval
def plot_f(x, F):
    plt.figure(figsize=(10,10))
    plt.plot(x, F, label="a Gaussian wave package")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Plot in arnbitary closed interval
X = np.linspace(0,1,1000)
plot_f(X, f(X))
```

```python
# plot f(x) in the range [−10, 10]
X = np.linspace(-10,10,1000)
plot_f(X, f(X))
```

#### part 2 ####

Analytical derivative of f(x).

```python
#Python function that calculates the derivative
def der_f(x, a=0.1, b=10):
    return b* np.cos(b*x)* np.exp(-a*(x**2))-2*a*x*np.sin(b*x)*np.exp(-a*(x**2))
```

#### part 3 ####

Function for the forward diffrence method.

```python
# calculates the derivative of an arbitrary function using the forward difference method
def fd(f,x,h):
    """ Evaluates the derivative of f(x) using forward difference method. h = step size"""
    return (f(x+h)-f(x))/h

print("analytical method:", der_f(1))
print("numerical method :",fd(f, 1, 1e-2))# x=1, h = 1e-2
```

#### part 4 ####

Function for the central differance method

```python
def fc(f,x,h):
    """ Evaluates the derivative of f(x) using central difference method, h = step size"""
    return 0.5*(f(x+h)-f(x-h))/h

print("central method :",fc(f, 1, 1e-2))
```

#### part 5 ####

Quantify the error in our numerical derivative
approximations

```python
## quantify the error in our numerical derivative approximations for the point x = 1
h = np.logspace(-16, 0, 100)
x = 1
plt.figure(figsize=(10,10))
plt.plot(h, abs(der_f(x)-fd(f,x,h)), "or", label="Forward difference")
plt.plot(h, abs(der_f(x)-fc(f,x,h)), "xb", label="Central difference")
plt.grid()
plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.show()
```

###  Exercise 3, Part II: FD with Classes ###

```python
class WavePacket:
    def __init__(self, a, b, x, h): #constructor
        self.a = a
        self.b = b
        self.x = x
        self.h = h
        
    def f(self): #function defined for wave phenomenon in python
        return np.sin(self.b*self.x)*np.exp(-self.a*self.x*self.x)
    
    def der_f(self): # analytical derivative of f(x)
        return self.b* np.cos(self.b*self.x)* np.exp(-self.a*(self.x**2))-2*self.a*self.x*np.sin(self.b*self.x)*np.exp(-self.a*(self.x**2))
    
    def fd(self):  # calculates the derivative of an arbitrary function using the forward difference method
        return (f(self.x+self.h)-f(self.x))/self.h
    
    def fc(self): # calculates the derivative of an arbitrary function using the central difference method
        return 0.5*(f(self.x+self.h)-f(self.x-self.h))/self.h

    def plot(self, x_min=-10, x_max=10, dx=0.01): # plot f(x) in the range [−10, 10]
        x = np.arange(x_min, x_max, dx)
        plt.figure(figsize=(10,10))
        plt.plot(x, f(x), label="a Gaussian wave package")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()
        plt.grid()
        plt.show()
        
    def scatter_error(self): #  scatter plots showing the absolute error of the two finite difference methods (central and forward) 
        plt.figure(figsize=(10,10))
        plt.plot(self.h, abs(der_f(self.x)-fd(f,self.x,self.h)), "or", label="Forward")
        plt.plot(self.h, abs(der_f(self.x)-fc(f,self.x,self.h)), "xb", label="Central")
        plt.grid()
        plt.legend()
        plt.xscale("log")
        plt.yscale("log")
        plt.show()

```

```python
h = np.logspace(-16, 0, 100)
x = 1
WP1 = WavePacket(0.1, 2, x, h) # a=0.1, b=2
WP1.plot()
WP1.scatter_error()

```

### Exercise 4: A song of ice and fire?


#### part 1 ####

Import glacier dataset from external source and plot the glacier. 

```python
import rockhound as rh
import cmocean
bedmap = rh.fetch_bedmap2(datasets=["thickness","surface","bed"])
plt.figure(figsize=(8, 7))
ax = plt.subplot(111)
bedmap.surface.plot.pcolormesh(ax=ax,cmap=cmocean.cm.ice,cbar_kwargs=dict(pad=0.01, aspect=30))
plt.title("Bedmap2 Antarctica")
plt.tight_layout()
plt.show()
```

#### part 2 ####

Slicing the data to get a 1d representation of an area on the glacier.

```python
# Extract cross section using the original coordinates
bed1d = bedmap.sel(y=-0.5e6, x=slice(-1.6e6,-1.35e6))
# Add a second x-coordinate to make prettier plots
# (shift x-axis to start at zero, and convert from m to km)
bed1d = bed1d.assign_coords({"x2":((bed1d.x+1.6e6)/1e3)})
```

```python
bed1d.surface.plot(x='x2') # plot surface elevation
```

```python
def Freebord_level(surface, thickness):
    p_w = 1.0 # density of water
    p_ice = 0.9340 # density of ice
    return -(surface-thickness)*((p_w/p_ice)-1) # Freeboard level
```

```python
surface = bed1d.surface.values # surface elevation
thickness = bed1d["thickness"].values # ice thickness
water = surface-thickness  # water depth
x = np.arange(0.0,len(bed1d.bed.values), 1) # x-axis

h = Freebord_level(surface, thickness)
plt.figure(figsize=(15,15))
plt.plot(h, "-r")
plt.plot(surface, "b")
plt.plot([0]*len(thickness), "--k")
plt.plot(water, "-k")
plt.plot(bed1d.bed.values, "k")
plt.fill_between(x, bed1d.bed.values, -2000, facecolor="k") # fill the bedrock with black
plt.fill_between(x, surface, bed1d.bed.values, facecolor="lightblue")  # fill the ice with lightblue
plt.fill_between(x, surface-thickness, bed1d.bed.values, facecolor="blue")  # fill the water with blue
plt.ylabel("hight above sea level [m]")
plt.xlabel("distance along track [km]")
plt.show()
```

From the figure, we can see that the first 40 km of data for the ocean is missing. After investigating, we realised this is because the dataset shows NaN values for that part. Apart from that, we can observe the bedrock in black color, The sea in deep blue color and the ice in light blue color. We can also observe freeboard level in a red line. 
We observe that the ice is thickest at over 1000 meters above sea level and thinnest at about some 10 meters above sea level.
 

Yes, the shape of the bedrock is important to sea level rise. Geologically speaking, it is necessary to recognise that when there exist accomodation space, sea level will generally be low however in instances where there is not, then we expect sea level to rise. With respect to this exercise, it can be observed that the ice thickness turns to increase along tracks. However, we can also observe that the bedrock of antarctica though undulating, generally turn to decrease in elevation along track. This inclination can create acomodation space for the melting ice to occupy once it melts away. Hence the shape of the bedrock plays a big role to the sea level rise and will therefore dictate the freeboard level. 

#### part 3 ####

calculating SLR if all the ice in antarctica chould melt.

```python
sea_A = 361*10**6 # km^2
Grid_cell = 1 * 1 #(km)
surface = bedmap.surface.values # surface elevation
thickness = bedmap.thickness.values # thickness of ice
bedrock = bedmap.bed.values # bed rock
```

```python
ASL = np.where(bedrock<0,0,bedrock) # extracting values for bedrock exposed above sea level
BSL = np.where(bedrock>0,0,bedrock) # extracting values for bedrock existing below the sea level
SBSL = np.where(BSL<0,surface,0) # extracting values for surface below sea level
TBSL= np.where(BSL<0,thickness,0) # extracting values for ice thickness below sea level
h = Freebord_level(SBSL, TBSL) # calculating the freeboard level
```

```python
TASL = np.where(ASL>0,thickness,0) # thickness of ice above sea level 
TSBSL = SBSL - h # thickness_of_ice_surface_to_freeboard or below sea level
total_thickness = TASL + TSBSL # Total thickness of ice
total_thickness_positive =np.where(total_thickness<0,0,total_thickness) # Total thickness of ice above sea level or total volume of ice above sea level
np.unique(total_thickness_positive)
```

```python
Total_volume = total_thickness_positive * Grid_cell 
Total_volume_grid = Total_volume[~np.isnan(Total_volume)] # remove nan values
Total_volume_grid = np.sum(Total_volume_grid)
print ('Total volume of ice that can cause sea level rise in (Km)',Total_volume_grid)
```

```python
SLR = Total_volume_grid / sea_A # Sea level rise
print ('Sea Level will rise',SLR,'m')
```
